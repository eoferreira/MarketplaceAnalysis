# Respostas do Desafio

```python
from google.cloud import bigquery
from IPython.core.display import display, HTML
from prettytable import PrettyTable
```

```python
def load_queries(filename):
    with open(filename, 'r') as file:    
        return file.read().split(';')
```

```python
filename = 'queries.sql'
modelo = 'Onix'
client = bigquery.Client.from_service_account_json('../../Downloads/ancient-medium-220620-dcb918ca109a.json')
query_list = load_queries(filename)
```

```python
# for query in query_list:
query_job = client.query("""
                         SELECT id, COUNT(id) 
                         FROM dataset_icarros.all_cars 
                         GROUP BY id 
                         HAVING COUNT(id) > 1 
                         LIMIT 10
                         """)
results = query_job.result()  # Waits for job to complete.
for row in results:
    print(row)
    query_job.to_dataframe().shape    
```

## Qual o preço médio do produto por cidade?

```python
query_job = client.query("""
                         SELECT cidade, UF, AVG(preco) as media
                         FROM dataset_icarros.all_cars 
                         WHERE modelo = '{}'
                         GROUP BY cidade, UF
                         ORDER BY cidade
                         """.format(modelo))
results = query_job.result()
t = PrettyTable(['Cidade', 'Preço médio'])
for row in results:
    t.add_row(['{} - {}'.format(row.cidade, row.UF), 'R$ {:,.2f}'.format(row.media).replace('.', '|').replace(',', '.').replace('|', ',')])
    
display(HTML('<h3>Preço médio do produto {} por cidade:</h3>'.format(modelo)))
print(t)
```

## Qual a cidade que possui mais ofertas do produto?

```python
query_job = client.query("""
                         SELECT cidade, UF, COUNT(id) as total
                         FROM dataset_icarros.all_cars 
                         WHERE modelo = '{}'
                         GROUP BY cidade, UF
                         ORDER BY total DESC
                         LIMIT 1
                         """.format(modelo))
results = query_job.result()
t = PrettyTable(['Cidade', 'Total'])
for row in results:
    t.add_row(['{} - {}'.format(row.cidade, row.UF), '{}'.format(row.total)])
display(HTML('<h3>Cidade com mais ofertas do produto {}:</h3>'.format(modelo)))
print(t)
```

## Liste os 5 mais baratos e o mais caro.

```python
query_job = client.query("""
                         SELECT versao, ano, km, cidade, UF, preco
                         FROM dataset_icarros.all_cars 
                         WHERE modelo = '{}'
                         ORDER BY preco ASC
                         LIMIT 5
                         """.format(modelo))
results = query_job.result()
t = PrettyTable(['Versão', 'Cidade', 'Ano', 'Km', 'Preço'])
print()
for row in results:
    t.add_row(['{}'.format(row.versao), 
               '{} - {}'.format(row.cidade, row.UF),
               '{}'.format(row.ano), 
               '{}'.format(row.km), 
               'R$ {:,.2f}'.format(row.preco).replace('.', '|').replace(',', '.').replace('|', ',')])
display(HTML('<h3>5 mais baratos:</h3>'))
print(t)
```

```python
query_job = client.query("""
                         SELECT versao, ano, km, cidade, UF, preco
                         FROM dataset_icarros.all_cars 
                         WHERE modelo = '{}'
                         ORDER BY preco DESC
                         LIMIT 1
                         """.format(modelo))
results = query_job.result()
t = PrettyTable(['Versão', 'Cidade', 'Ano', 'Km', 'Preço'])
print()
for row in results:
    t.add_row(['{}'.format(row.versao), 
               '{} - {}'.format(row.cidade, row.UF),
               '{}'.format(row.ano), 
               '{}'.format(row.km), 
               'R$ {:,.2f}'.format(row.preco).replace('.', '|').replace(',', '.').replace('|', ',')])
display(HTML('<h3>O mais caro:</h3>'))
print(t)
```

```python
query_job = client.query("""
                         SELECT COUNT (*) as total FROM (
                             SELECT DISTINCT cidade, UF
                             FROM dataset_icarros.all_cars 
                             WHERE modelo = '{}'
                                 AND CIDADE IS NOT NULL 
                                 AND CIDADE <> ''
                             GROUP BY cidade, UF
                             ORDER BY cidade
                         )
                         """.format(modelo))
results = query_job.result()  # Waits for job to complete.
# print(results.result())
t = PrettyTable(['Total de Cidades'])
for row in results:
    t.add_row(['{}'.format(row.total)])
display(HTML('<h4>Quantas cidades possuem anúncios do produto {}?</h4>'.format(modelo)))    
print(t)
```

```python
query_job = client.query("""
                         SELECT COUNT (*) as total FROM (
                             SELECT DISTINCT UF
                             FROM dataset_icarros.all_cars 
                             WHERE modelo = '{}'
                                 AND UF IS NOT NULL 
                                 AND UF <> ''
                             GROUP BY UF
                             ORDER BY UF
                         )
                         """.format(modelo))
results = query_job.result()
t = PrettyTable(['Total de Estados'])
for row in results:
    t.add_row(['{}'.format(row.total)])
display(HTML('<h4>Quantos estados possuem anúncios do produto {}?</h4>'.format(modelo)))    
print(t)
```

```python
query_job = client.query("""
                         SELECT cidade, UF, AVG(preco) as media
                         FROM dataset_icarros.all_cars 
                         GROUP BY cidade, UF
                         ORDER BY cidade
                         """)
results = query_job.result()
t = PrettyTable(['Cidade', 'Preço médio'])
for row in results:
    if row.cidade == '':
        cidade = 'N/D'
    else:
        cidade = row.cidade
    t.add_row(['{} - {}'.format(cidade, row.UF), 'R$ {:,.2f}'.format(row.media).replace('.', '|').replace(',', '.').replace('|', ',')])
    
display(HTML('<h3>Preço médio do produto {} por cidade:</h3>'.format(modelo2)))
print(t)
```
